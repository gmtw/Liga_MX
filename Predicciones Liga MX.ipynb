# A partir de ahora creamos un algoritmo para web scrapping

```python
import pandas as pd #Traemos las librerias
import numpy as np
from bs4 import BeautifulSoup  #Biblioteca para web scrapring
from urllib.request import urlopen #Biblioteca para abrir urls
```

```python
def scrapping_ligaMX(url):
   # import pandas as pd #Traemos las librerias
   # import numpy as np
   # from bs4 import BeautifulSoup  #Biblioteca para web scrapring
   # from urllib.request import urlopen #Biblioteca para abrir urls
    
    pagina = urlopen(url) #Abrimos la página web que queremos sacar la información
    
    soup = BeautifulSoup(pagina.read(),"html5lib") #Creamos un objeto donde vamos a guardar nuestra página
    
    tabla_general = soup.find('table', class_='default tbl_grals') #Variable que guarda la tabla general <tr> de
    #liga MX
    # obviamente esta escrito en html
    
    tag_tr=tabla_general.find_all('tr') #Esta variable guarda todas las etiquetas <tr>
    #Las etiquetas <tr> define una fila de celdas en una tabla. 
    #Estas pueden ser una mezcla de elementos <td> y <th>.
    
    lista_columnas = columnas(tag_tr=tag_tr,posicion=1) #obtiene los nombres de la columna de nuestra tabla
    
    #Listas vacias para obtener el contenido de nuestra tabla
    
    posiciones_club = [] 
    nombres_club = []
    datos_club = [] #Datos tales como juegos jugados, ganados , perdidos etc
    
    for equipos in range(2, len(tag_tr)): #de 2 al numero de equipos
        posiciones_club.append(posicion(tag_tr,equipos))
        nombres_club.append(nombre(tag_tr, equipos))
        datos_club.append(datos(tag_tr, equipos))
        
    posiciones_club = np.array(posiciones_club)
    nombres_club = np.array(nombres_club)
    datos_club = np.array(datos_club)
    
    
    
    
    
    df1 = pd.DataFrame(datos_club)
    df2 = pd.DataFrame(nombres_club)
    df3 = pd.DataFrame(posiciones_club)
    
    df = pd.concat([df3, df2, df1], axis=1)
    df.columns = columnas(tag_tr)
    

    
    #print(posiciones_club.reshape(18,1), nombres_club.reshape(18,1),datos_club)
    #print(np.array([posiciones_club.reshape(18,1), nombres_club.reshape(18,1),datos_club.reshape(18,1)]))
    
    #tabla = pd.DataFrame(pre_tabla)

    
    return df
    
    
        
    
```

```python
def columnas(tag_tr,posicion=1):
    #default posicion=1
    cabecera=[text for text in tag_tr[posicion].stripped_strings] #El método stripped sirve para limpiar las
    #cadenas de elementos no deseados
    cabecera = np.array(cabecera)
    
    return cabecera 
    
```

```python
def posicion(tag_tr,i):
    contenido =tag_tr[i].find_all('td')[0].contents # obtenemos el contenido del tag <td>
    pos = int(contenido[0].rstrip('\n\t')) # limpiamos el contenido de saltos de linea y tabulaciones
    
    return pos    
```

```python
def nombre(tag_tr, i):
    equipos_html = etiquetas_tr[i].find('a', class_='tpts') #etiqueta donde se encuentra el equipo
    contenido = equipos_html.contents #Contenido de la etiqueta
    title_team= contenido[0].rstrip('\n\t').lstrip('\n ') #limpieza de la etiqueta
    
    return title_team
```

```python
def datos(tag_tr, i):
    
    data_prin = []
    
    tamaño_columna = len(etiquetas_tr[i].find_all('td'))
    inicio = 2
    j=inicio

    for column in range(inicio,tamaño_columna):
        
        if len(str(etiquetas_tr[i].find_all('td')[column].contents[0])) < 5:

            dato = etiquetas_tr[i].find_all('td')[column].contents[0]
            

        else:

            dato = etiquetas_tr[i].find_all('a', class_='loadershow')[j].contents[0]
            j=j+1
        
        data_prin.append(dato)

        #print(column,j)
        #print(dato)


    return data_prin
```
