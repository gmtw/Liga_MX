# A partir de ahora vamos a convertir a una clase

```python
class Tabla():
    
    def __init__(self, url='https://ligamx.net/cancha/estadisticahistorica',equipos=19):
       
        self.url = url
        self.equipos = equipos
    def sopa(self):
        
        html = urlopen(self.url) #Pagina de la Liga MX en forma de url
        
        soup = BeautifulSoup(html.read(),"html5lib") #Importamos el texto del código fuente
        
        #============escudriño========================================================
        
        tabla_general = soup.find('table', class_='default tbl_grals') 
        
        self.etiquetas_tr=tabla_general.find_all('tr') #Las celdas que van dentro de cada fila
#las tenemos que escribirlas con la etiqueta <td> y su correspondiente cierre </td>.
        
        return self
    
    def etiquetas_tr(self):
        self.etiqueta = self.sopa().etiquetas_tr
        return self.etiqueta
        
        
    def cabecera(self):
        #a = self.etiquetas_tr[1]
        cabecera=[text for text in self.etiquetas_tr[1].stripped_strings] #Stripped_strings obtine la cadena de la etiqueta
        return cabecera
    
    def posiciones(self):
        posicion = [int(self.etiquetas_tr[pos].find_all('td')[0].contents[0].rstrip('\n\t\t')) for pos in range(2,self.equipos+2)]
        data_posicion = pd.DataFrame(posicion)
        return data_posicion
    
    def clubes(self):
        
        club = [self.etiquetas_tr[nombre].find('a', class_='tpts').contents[0].rstrip('\n\t').lstrip('\n ') for nombre in range(2,self.equipos+2)]
        data_club = pd.DataFrame(club)
        return data_club
    
    def caracteristica(self):
        caracteristicas = []
        
        for row in range(2,self.equipos+2): #20 es el número de equipos
            car = []
            for column in range(2,26): #25 es el numero de características
                for child in self.etiquetas_tr[row].find_all('td')[column].strings:
                    car.append(int(child))

            caracteristicas.append(car)
        
        return pd.DataFrame(caracteristicas)
    
    def tabla(self):
        tabla_data = pd.concat([self.posiciones(),self.clubes(), self.caracteristica()],axis = 1)
        tabla_data.columns = self.cabecera()

        return tabla_data
```
